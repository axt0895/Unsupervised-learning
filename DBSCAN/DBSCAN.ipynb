```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_moons
from mlxtend.plotting import plot_decision_regions
from sklearn.cluster import DBSCAN
```

```python
X, y = make_moons(n_samples=600, random_state=42, noise=0.11, shuffle=True)
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(X, y)
```

```python
fig = plot_decision_regions(X, y, clf= dbscan, legend=1)
```
