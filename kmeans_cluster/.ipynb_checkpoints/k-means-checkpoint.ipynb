# Imports

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from mlxtend.plotting import plot_decision_regions
```

# Datasets

```python
X, y = make_blobs(n_samples=800, n_features=2, centers=6, random_state=42, shuffle=True)
```

# Plotting the datasets

```python
plt.scatter(X[:, 0], X[:, 1], s = 10)
```

# KMeans Algorithm

```python
k = 5 
kmeans = KMeans(n_clusters=5, n_init=10, random_state=42)
```

```python
y_predict = kmeans.fit_predict(X)
```

# Labels of the fittted instances

```python
y_predict
```

# Cluster Centriod

```python
kmeans.cluster_centers_
```

# Assigning cluster to new centriod

```python
X_new = np.array([[0, 2], [3, 2],[-3, 3],[-3, 4.5]])
```

```python
kmeans.predict(X_new)
```

# Visualize the decision boundary

```python
plt.figure(figsize=(8, 6))
fig = plot_decision_regions(X, y, clf=kmeans)
```

# Soft Voting classifier -- Assigning number of each instance rather than hard classifier

```python
kmeans.transform(X_new).round(1)
```
