```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from mlxtend.plotting import plot_decision_regions
```

```python
X, y = make_blobs(n_samples=200, n_features=2, centers=5,random_state=42, shuffle=True)
```

```python
plt.scatter(X[:,0], X[:, 1], s= 10)
```

```python
k=5
k_means = KMeans(n_clusters=k, random_state=42, n_init='auto')
```

```python
y_predict = k_means.fit_predict(X)
```

```python
k_means.cluster_centers_
```

```python
import numpy as np
```

```python
X_new = np.array([[0, 2],[3, 2],[-3,3],[-3,2.5]])
```

```python
k_means.predict(X_new)
```

```python
plt.figure(figsize=(8, 6))
fig = plot_decision_regions(X, y, clf=k_means)
```
